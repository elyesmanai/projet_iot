```python
print(df.shape)
# Drop columns with a single unique value
df = df.loc[:, df.nunique() > 1]

print(df.shape)  # We dropped 164 columns
```

```python
df.dropna(how="all", axis=0)
# 0 lignes  où toutes les valeurs sont nulles
# 0 colonnes où toutes les valeurs sont nulles
```

```python
unnamed = [col for col in df.columns if 'Unnamed' in col]

for feature in unnamed:
    print(df[feature].value_counts().sum())

# le nombre de valeurs existantes par raport au nombre de lignes totales est trop petit (max 74/211043) donc on va enlever toutes les colonnes unnamed
```

```python
print(df.shape)
df.drop(unnamed, axis=1, inplace=True)
print(df.shape)
# il y avait 84 colonnes unnamed qui avaient quand meme 2 valeurs ou plus
```

```python
df['type'].unique() # types 1 - 0 37 WOW sql...  total 74 vals sur 210k, drop

df = df[~df['type'].isin(['1', '0', '-', ' WOW64)', 'sqlmap/1.2#stable (http://sqlmap.org)', '37'])]
df['type'].unique()


df['label'] = df['label'].astype('int')
```

```python
df.groupby('label')['type'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = le.fit_transform(df[col].astype(str))
```
